## 📖 Part A – Theoretical Foundation (Short Notes)


### ❓ Q1. What is a Statistical Distribution?

**Answer:**

A **statistical distribution** simply tells you how the values of something are spread out. It shows which values show up a lot, which show up rarely, and how everything is arranged from the smallest to the biggest value.

Distributions come in two main types:
- **Discrete** — you can count the outcomes (like how many transactions happen in a day: 0, 1, 2, 3...)
- **Continuous** — the values can be anything in a range, including decimals (like a transaction amount in ₹)

**Why it matters:** once we know the shape of the data, we can guess how likely future values are — which is exactly what businesses need for planning and risk.

| Property | What it tells us |
|---|---|
| Central tendency | Where most values sit (mean, median, mode) |
| Spread | How wide the values are scattered (variance, std dev) |
| Shape | Symmetric, skewed, peaked, flat, etc. |

**Simple example:** most online purchases are small to medium, but every now and then someone makes a really big purchase. That long tail of big purchases is what we call a right-skewed distribution.

### ❓ Q2. What is a Q-Q Plot and why is it used?

**Answer:**

A **Q-Q Plot (Quantile-Quantile Plot)** is a simple way to check if your data looks like a Normal (bell-curve) distribution. It lines up your actual data against what a perfect Normal distribution would look like, point by point.

**How to read it:**

```
Points sit close to the straight line  -> data is close to Normal
Points curve away at the ends          -> tails are heavier/lighter than expected
Points bend upward overall             -> data is right-skewed
```

**Why it's useful:** a histogram can hide small but important differences, especially at the extreme ends. A Q-Q plot makes those differences much easier to spot just by looking.

| Pattern | What it means |
|---|---|
| Straight line | Data is roughly Normal |
| Curves upward | Right-skewed data |
| Curves downward | Left-skewed data |
| S-shape | Heavier or lighter tails than Normal |

**Simple example:** before testing if average transaction amounts differ between two groups, we first check a Q-Q plot to see if the Normal-distribution assumption actually makes sense for our data.


*Fig 2: Q-Q plot of well-behaved Normal data (left) vs right-skewed data (right) — note the curve at the tail*



### ❓ Q3. Difference between Discrete and Continuous Distributions.

**Answer:**

A **discrete distribution** is for things you can count one by one — there are gaps between the possible values (0, 1, 2, 3...). A **continuous distribution** is for things that can take any value, including fractions and decimals, within a range.

**Quick way to tell them apart:** if you can list every possible outcome, it's discrete. If the outcomes flow smoothly without gaps, it's continuous.

| Aspect | Discrete | Continuous |
|---|---|---|
| Values | Countable (0,1,2,...) | Any number in a range |
| Probability function | PMF (Probability Mass Function) | PDF (Probability Density Function) |
| Probability of one exact value | Can be greater than 0 | Always 0 (we measure ranges instead) |
| Examples | Bernoulli, Binomial, Poisson | Normal, Log-Normal, Power Law |
| Dataset example | transaction_count (per week) | transaction_amount (₹) |

**Simple example:** how many times a customer shops per week (0, 1, 2...) is discrete. How much money they spend (₹2,781.84) is continuous.


*Fig 3: Discrete PMF (bars, countable) vs Continuous PDF (smooth curve)*



### ❓ Q4. What is Bernoulli Distribution?

**Answer:**

The **Bernoulli distribution** is the simplest distribution there is — it's just one trial with two possible results: success (1) or failure (0). The probability of success is called `p`. It's basically a single coin flip, where the coin doesn't have to be fair.

**Formula:**
```
P(X=1) = p
P(X=0) = 1 - p
Mean = p
Variance = p(1-p)
```

| Outcome | Probability |
|---|---|
| Success (1) | p |
| Failure (0) | 1 - p |

**Simple example:** every single transaction is either `Success` or `Fail` — that's a Bernoulli trial in action.


*Fig 4: Bernoulli distribution for a transaction Success/Fail outcome*



### ❓ Q5. What is Binomial Distribution?

**Answer:**

The **Binomial distribution** is what you get when you repeat a Bernoulli trial `n` times and count how many successes you got. It answers a simple question: "out of n tries, how many will succeed?"

**Formula:**
```
P(X=k) = C(n,k) * p^k * (1-p)^(n-k)
Mean = n * p
Variance = n * p * (1-p)
```

| Parameter | Meaning |
|---|---|
| n | Number of tries |
| p | Chance of success on each try |
| k | Number of successes we observed |

**Simple example:** if a customer makes several transaction attempts in a week, the number of those that succeed follows a Binomial distribution.


*Fig 5: Binomial PMF — successes out of n weekly transaction attempts*



### ❓ Q6. Explain Log-Normal Distribution.

**Answer:**

A variable follows a **Log-Normal distribution** when taking its logarithm gives you something that looks Normal. In practice, this shows up as a shape that's **right-skewed** — lots of small/medium values with a long tail of bigger ones. It happens whenever a value can never go below zero and tends to grow by multiplying rather than adding.

**Formula (PDF):**
```
f(x) = 1 / (x * sigma * sqrt(2*pi)) * exp( -(ln(x) - mu)^2 / (2*sigma^2) ),  x > 0
```

| Property | Behavior |
|---|---|
| Domain | x > 0 only (never negative) |
| Shape | Right-skewed, long tail |
| Common use | Income, transaction sizes, stock prices, file sizes |

**Simple example:** most transaction amounts on an e-commerce platform are modest, but a few large purchases stretch the tail out to the right — that's a classic Log-Normal pattern.


*Fig 6: Log-Normal distribution shapes — higher sigma means a longer right tail*



### ❓ Q7. Explain Power Law Distribution.

**Answer:**

A **Power Law distribution** is for cases where small values are extremely common and large values are rare — but not as rare as you'd expect. The tail is even heavier than a Log-Normal distribution's tail.

**Formula:**
```
P(X = x) ~ x^(-alpha),   x >= x_min
```

| Property | Behavior |
|---|---|
| Shape | Very heavy ("fat") tail |
| Key feature | Looks similar whether you zoom in or out |
| Common use | City populations, word frequencies, wealth distribution, viral content |

**Simple way to think about it:** a small number of extreme values can end up contributing way more to the total than a Normal or Log-Normal model would suggest.

**Simple example:** if just a handful of "whale" customers account for most of the total transaction value, a Power Law might describe their spending better than a Log-Normal would.


*Fig 7: Power Law on a log-log scale — a straight line is the signature of a true power law*



### ❓ Q8. What is Box-Cox Transform?

**Answer:**

The **Box-Cox transform** is a trick used to take skewed data and reshape it to look more Normal. This helps a lot of statistical methods work properly, since many of them assume the data is roughly Normal to begin with.

**Formula:**
```
y(lambda) = (x^lambda - 1) / lambda   if lambda != 0
y(lambda) = ln(x)                      if lambda = 0
```

The best `lambda` value is picked automatically so the transformed data ends up as close to Normal as possible. One rule: every value has to be positive for this to work.

| Lambda value | Effect |
|---|---|
| lambda = 1 | No change |
| lambda = 0 | Same as a log transform |
| lambda = 0.5 | Same as a square-root transform |
| lambda < 0 | Flips and shrinks large values |

**Simple example:** before running a t-test or building a model that expects Normal data, we can Box-Cox transform the skewed transaction amounts first.


*Fig 8: Box-Cox transform reduces skew and brings the distribution closer to Normal*



### ❓ Q9. Explain Poisson Distribution with an example.

**Answer:**

The **Poisson distribution** is used to count how many times something happens in a fixed window of time or space, when events happen independently and at a steady average rate (`lambda`).

**Formula:**
```
P(X=k) = (lambda^k * e^(-lambda)) / k!
Mean = lambda
Variance = lambda
```

| Parameter | Meaning |
|---|---|
| lambda | Average number of events in that window |
| k | The actual count we observed |

**Simple example:** the number of transactions a customer makes per day, assuming each one happens independently at roughly the same average rate, fits well with a Poisson distribution where lambda is the average daily count.


*Fig 9: Poisson distribution — higher lambda shifts and spreads the peak rightward*



### ❓ Q10. What is Z-score Probability?

**Answer:**

A **Z-score** tells you how far a value is from the average, measured in standard deviations. Once you have the Z-score, you can use the standard Normal distribution to find the probability of seeing a value that extreme or more.

**Formula:**
```
Z = (x - mean) / std_dev
P(X > x) = 1 - Phi(Z)        where Phi = standard normal CDF
```

| Z-score | What it means |
|---|---|
| Z = 0 | Right at the average |
| Z = +1 | One standard deviation above average |
| \|Z\| > 2 | Pretty far from average (roughly top/bottom 5%) |
| \|Z\| > 3 | Very unusual (roughly top/bottom 0.3%) |

**Simple example:** turning a transaction amount into a Z-score lets us estimate the chance that a random transaction goes over ₹5,000, assuming the data is Normal.


*Fig 10: The shaded area is the probability of exceeding a given Z-score threshold*



### ❓ Q11. Differentiate Probability Density Function (PDF) and Cumulative Distribution Function (CDF).

**Answer:**

The **PDF** shows how likely values are near a certain point — it's the height of the curve, not a probability by itself. The **CDF** tells you the probability that the variable is **less than or equal to** a given value — it's basically a running total of the PDF.

**Formula:**
```
CDF(x) = P(X <= x) = integral of PDF from -infinity to x
PDF(x) = derivative of CDF(x)
```

| | PDF | CDF |
|---|---|---|
| Shows | The shape/density at each point | The total probability built up so far |
| Range | Can go above 1 (it's a density, not a probability) | Always between 0 and 1 |
| Always increasing? | No | Yes, it never goes down |
| Used for | Seeing the shape of the data | Reading off probabilities directly |

**Simple example:** the PDF shows where most transaction amounts cluster. The CDF directly answers "what percent of transactions are under ₹3,000?"
